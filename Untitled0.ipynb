```python
# MAMI-2022 Misogynous Memes Analysis
# This script performs analyses on the MAMI-2022 dataset:
# 1. Class distribution
# 2. Sexist lexicon counts
# 3. TF-IDF n-grams (uni- & bi-grams)
# 4. Face–gender ratio (DeepFace)
# 5. CLIP image–text similarity
#
# Paths and column names are hardcoded for convenience.

import os
import re
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud

# Attempt DeepFace import
try:
    from deepface import DeepFace
    DEEPFACE_AVAILABLE = True
except ImportError:
    DEEPFACE_AVAILABLE = False
    print("Warning: deepface module not found. Face–gender analysis will be skipped.")

import clip
import torch
from PIL import Image

# Hardcoded paths
CSV_PATH = "/content/drive/MyDrive/MAMI2022/train.tsv"
IMG_DIR  = "/content/drive/MyDrive/MAMI2022/MAMI_2022_images/training_images"

# Hardcoded column names
IMG_COL   = 'file_name'
TEXT_COL  = 'text'
LABEL_COL = 'label'


def load_mami_data(path: str) -> pd.DataFrame:
    """
    Load the MAMI file into a DataFrame.
    Uses tab-delimiter for .tsv or retries comma.
    """
    ext = os.path.splitext(path)[1].lower()
    if ext in ['.tsv', '.txt']:
        df = pd.read_csv(path, sep='\t', engine='python', on_bad_lines='skip')
    else:
        try:
            df = pd.read_csv(path)
        except pd.errors.ParserError:
            print("Warning: failed comma parse; retrying tab-delimited...")
            df = pd.read_csv(path, sep='\t', engine='python', on_bad_lines='skip')
    return df


def class_distribution(df: pd.DataFrame) -> None:
    counts = df[LABEL_COL].value_counts().sort_index()
    plt.figure(figsize=(6,4))
    counts.plot(kind='bar', color=['grey','red'])
    plt.xticks([0,1], ["Benign","Misogynous"])
    plt.title('Class Distribution')
    plt.ylabel('Count')
    plt.xlabel('Label')
    plt.tight_layout()
    plt.show()


def sexist_lexicon_analysis(df: pd.DataFrame) -> None:
    slurs = ['bitch','slut','whore','cunt','tits','pussy']
    for s in slurs:
        df[f'count_{s}'] = df[TEXT_COL].astype(str).str.lower().apply(lambda t: len(re.findall(r"\b"+s+r"\b", t)))
    agg = df.groupby(LABEL_COL)[[f'count_{s}' for s in slurs]].mean()
    agg.index = ["Benign","Misogynous"]
    agg.plot(kind='bar', figsize=(8,5))
    plt.title('Average Sexist Slur Count by Class')
    plt.ylabel('Average Count')
    plt.xlabel('Label')
    plt.tight_layout()
    plt.show()


def tfidf_ngram_analysis(df: pd.DataFrame) -> None:
    """
    Compute TF-IDF for each class and visualize top 20 uni/bi-grams.
    """
    vec = TfidfVectorizer(ngram_range=(1,2), stop_words='english', max_features=5000)
    X = vec.fit_transform(df[TEXT_COL].fillna(''))
    feats = vec.get_feature_names_out()
    classes = sorted(df[LABEL_COL].unique())

    for cls in classes:
        mask = (df[LABEL_COL] == cls).values
        X_cls = X[mask]
        scores = X_cls.mean(axis=0).A1
        idx = scores.argsort()[::-1][:20]
        top = {feats[i]: scores[i] for i in idx}
        title = 'Misogynous' if cls == 1 else 'Benign'
        print(f"Top 20 n-grams for {title}:")
        for term, score in top.items():
            print(f"  {term}: {score:.4f}")
        wc = WordCloud(width=800, height=400).generate_from_frequencies(top)
        plt.figure(figsize=(10,5))
        plt.imshow(wc, interpolation='bilinear')
        plt.axis('off')
        plt.title(f"Word Cloud ({title})")
        plt.show()


def face_gender_ratio(df: pd.DataFrame) -> None:
    """
    Use DeepFace to detect face & gender; plot proportions per class.
    Handles missing image files gracefully.
    """
    if not DEEPFACE_AVAILABLE:
        print('Skipping face–gender ratio; deepface not installed.')
        return

    results = []
    for cls in sorted(df[LABEL_COL].unique()):
        subset = df[df[LABEL_COL] == cls]
        counts = {'no_face': 0, 'male': 0, 'female': 0}
        for fname in subset[IMG_COL]:
            img_path = os.path.join(IMG_DIR, fname)
            if not os.path.exists(img_path):
                print(f"Warning: image file not found: {img_path}")
                counts['no_face'] += 1
                continue
            try:
                face_img = DeepFace.detectFace(img_path, detector_backend='mtcnn')
                analysis = DeepFace.analyze(face_img, actions=['gender'], enforce_detection=False)
                g = analysis.get('gender', '')
                if g == 'Man':
                    counts['male'] += 1
                else:
                    counts['female'] += 1
            except Exception:
                counts['no_face'] += 1
        total = sum(counts.values())
        results.append({'Class': cls, **{k: (v/total if total>0 else 0) for k, v in counts.items()}})

    plot_df = pd.DataFrame(results).set_index('Class')
    plot_df.plot(kind='bar', stacked=True, figsize=(7,5))
    plt.title('Face–Gender Proportion per Class')
    plt.ylabel('Proportion')
    plt.tight_layout()
    plt.show()


def clip_congruence(df: pd.DataFrame) -> None:
    """
    Compute CLIP cosine similarity between images & text; boxplot by class.
    Handles missing image files gracefully.
    """
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    model, preprocess = clip.load('ViT-B/32', device=device)
    sims = []
    labels = []

    for _, row in df.iterrows():
        img_path = os.path.join(IMG_DIR, row[IMG_COL])
        if not os.path.exists(img_path):
            print(f"Warning: image file not found: {img_path}")
            continue
        try:
            img_tensor = preprocess(Image.open(img_path)).unsqueeze(0).to(device)
            txt_tensor = clip.tokenize([row[TEXT_COL] or '']).to(device)
            with torch.no_grad():
                img_emb = model.encode_image(img_tensor)
                txt_emb = model.encode_text(txt_tensor)
            sims.append(torch.cosine_similarity(img_emb, txt_emb).item())
            labels.append(row[LABEL_COL])
        except Exception as e:
            print(f"Error processing {img_path}: {e}")
            continue

    if not sims:
        print('No CLIP similarities computed.')
        return

    df_sim = pd.DataFrame({'clip_similarity': sims, 'label': labels})
    df_sim.boxplot(column='clip_similarity', by='label', figsize=(6,5))
    plt.title('CLIP Similarity')
    plt.suptitle('')
    plt.ylabel('Cosine Similarity')
    plt.xlabel('Class')
    plt.tight_layout()
    plt.show()


if __name__ == '__main__':
    df = load_mami_data(CSV_PATH)
    class_distribution(df)
    sexist_lexicon_analysis(df)
    tfidf_ngram_analysis(df)
    face_gender_ratio(df)
    clip_congruence(df)
    print('Analysis complete.')
```
